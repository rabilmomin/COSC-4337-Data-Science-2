```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns


from tensorflow.keras.utils import image_dataset_from_directory
```

```python
DATADIR = "/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery"
img_height = 64
img_width = 64
batch_size = 32
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot(3, 2, 1)
img = mpimg.imread("/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery/Trees/Forest_1549.jpg")
plt.imshow(img)
plt.title("Trees")
plt.xlabel("Forest_1549.jpg")
plt.xticks([])
plt.yticks([])
ax = plt.subplot(3, 2, 3)
img = mpimg.imread("/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery/Trees/River_57.jpg")
plt.imshow(img)
plt.title("Trees")
plt.xlabel("River_57.jpg")
plt.xticks([])
plt.yticks([])
ax = plt.subplot(3, 2, 5)
img = mpimg.imread("/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery/Trees/Pasture_1146.jpg")
plt.imshow(img)
plt.title("Trees")
plt.xlabel("Pasture_1146.jpg")
plt.xticks([])
plt.yticks([])
ax = plt.subplot(3, 2, 2)
img = mpimg.imread("/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery/NoTrees/AnnualCrop_30.jpg")
plt.imshow(img)
plt.title("NoTrees")
plt.xlabel("AnnualCrop_30.jpg")
plt.xticks([])
plt.yticks([])
ax = plt.subplot(3, 2, 4)
img = mpimg.imread("/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery/NoTrees/Industrial_36.jpg")
plt.imshow(img)
plt.title("NoTrees")
plt.xlabel("Industrial_36.jpg")
plt.xticks([])
plt.yticks([])
ax = plt.subplot(3, 2, 6)
img = mpimg.imread("/content/Trees_in_Satellite_Imagery/Trees in Satellite Imagery/NoTrees/SeaLake_1897.jpg")
plt.imshow(img)
plt.title("NoTrees")
plt.xlabel("SeaLake_1897.jpg")
plt.xticks([])
plt.yticks([])
```

```python
train_ds = image_dataset_from_directory(
    DATADIR,
    validation_split = 0.3,
    subset = "training",
    seed = 123,
    image_size = (img_height, img_width),
    batch_size = batch_size)
```

```python
val_ds = image_dataset_from_directory(
    DATADIR,
    validation_split = 0.3,
    subset = "validation",
    seed = 123,
    image_size = (img_height, img_width),
    batch_size = batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
train_x = []
train_y = []
for images, labels in train_ds:
  for i in range(images.shape[0]):
    train_x.append(images[i].numpy())
    train_y.append(labels[i].numpy())

  
train_x = np.array(train_x)
train_y = np.array(train_y)
print(train_x.shape)
print(train_y.shape)
train_x = train_x / 255.0
train_x_shaped = train_x.reshape(train_x.shape[0], img_height * img_width * 3)
```

```python
print(train_x_shaped)
```

```python
test_x = []
test_y = []
for images, labels in val_ds:
  for i in range(images.shape[0]):
    test_x.append(images[i].numpy())
    test_y.append(labels[i].numpy())
test_x = np.array(test_x)
test_y = np.array(test_y)
print(test_x.shape)
print(test_y.shape)
test_x = test_x / 255.0
test_x_shaped = test_x.reshape(test_x.shape[0], img_height * img_width * 3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators = 50, verbose = 2)
```

```python
model.fit(train_x_shaped,train_y)
```

```python
from sklearn import tree
tree.plot_tree(model[0])
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
pred_y=model.predict(test_x_shaped)
pred_y
```

```python
print(accuracy_score(pred_y, test_y))
print("Testing Classification Report")
print(classification_report(pred_y, test_y))
```

```python
model_cm = confusion_matrix(test_y, pred_y)
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5
ax = sns.heatmap(model_cm, annot = True, cmap = 'Blues', fmt = 'g')

ax.set_title('Initial Random Forest Confusion Matrix')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
ax.xaxis.set_ticklabels(['NoTrees', 'Trees'])
ax.yaxis.set_ticklabels(['NoTrees', 'Trees'])
plt.show()
```

```python
train_scores = []
test_scores = []
num_trees = []
for k in range(5, 25, 5):
  num_trees.append(k)
  model=RandomForestClassifier(n_estimators = k, verbose = 1, max_depth = 10)
  model.fit(train_x_shaped,train_y)
  pred_y=model.predict(train_x_shaped)
  train_scores.append(accuracy_score(pred_y, train_y))
  pred_y=model.predict(test_x_shaped)
  test_scores.append(accuracy_score(pred_y, test_y))
```

```python
rcParams['figure.figsize'] = 6, 4
plt.plot(num_trees, train_scores, label='Training Accuracy')
plt.plot(num_trees, test_scores, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.xlabel("Number of Trees")
plt.title('Training and Validation Accuracy')
```
